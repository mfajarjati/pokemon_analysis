#Exploratory data analysis

###1.mengisi kekosongan data

```python
# Mengisi nilai yang hilang pada kolom 'Type 2' dengan 'Unknown'
df['Type 2'].fillna('Unknown', inplace=True)
df
```

```python
# cek kekosongan (null) kembali dari dataset
df.isnull().sum()
```

###2. Menghapus kolom #



```python
# Menghapus kolom '#'
df = df.drop('#', axis=1)

# Mereset index dan menjadikannya kolom 'num'
df = df.reset_index()

# Mengganti nama kolom 'index' menjadi 'num'
df = df.rename(columns={'index': 'num'})

# Menambahkan 1 ke kolom 'num' agar dimulai dari 1
df['num'] = df['num'] + 1

# Mengatur kolom 'num' sebagai index
df.set_index('num', inplace=True)

df
```

```python
df.describe()
```

#data analysis

###Mengidentifikasi Distribusi Statistik

```python
import matplotlib.pyplot as plt
import seaborn as sns

stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
for stat in stats:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[stat], kde=True, bins=30)
    plt.title(f'Distribusi {stat}')
    plt.xlabel(stat)
    plt.ylabel('Frekuensi')
    plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(18, 12))
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

for i, stat in enumerate(stats):
    sns.histplot(df[stat], kde=True, bins=30, ax=axs[i//3, i%3])
    axs[i//3, i%3].set_title(f'Distribusi {stat}')
    axs[i//3, i%3].set_xlabel(stat)
    axs[i//3, i%3].set_ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

###Membandingkan Jenis Pokemon

```python
df.info()
```

```python
# Pastikan kolom numerik hanya mengandung nilai numerik
numerik_kolom = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total']
for kolom in numerik_kolom:
    df[kolom] = pd.to_numeric(df[kolom], errors='coerce')

# Hapus baris yang mengandung nilai NaN setelah konversi ke numerik
df = df.dropna(subset=numerik_kolom)

# Membuat subplots 3x2 untuk perbandingan statistik dasar berdasarkan Type 1
fig, axs = plt.subplots(2, 3, figsize=(18, 12))

stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
for i, stat in enumerate(stats):
    sns.boxplot(x='Type 1', y=stat, data=df, ax=axs[i//3, i%3])
    axs[i//3, i%3].set_title(f'Perbandingan {stat} berdasarkan Type 1')
    axs[i//3, i%3].set_xlabel('Type 1')
    axs[i//3, i%3].set_ylabel(stat)
    axs[i//3, i%3].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

###Mengetahui Pengaruh Generasi

```python
# Membuat subplots 3x2 untuk perbandingan statistik dasar berdasarkan Generation
fig, axs = plt.subplots(2, 3, figsize=(18, 12))

for i, stat in enumerate(stats):
    sns.boxplot(x='Generation', y=stat, data=df, ax=axs[i//3, i%3])
    axs[i//3, i%3].set_title(f'Perbandingan {stat} berdasarkan Generation')
    axs[i//3, i%3].set_xlabel('Generation')
    axs[i//3, i%3].set_ylabel(stat)

plt.tight_layout()
plt.show()

# Line plot untuk melihat tren statistik dasar dari generasi ke generasi
mean_stats_by_gen = df.groupby('Generation')[stats].mean()
plt.figure(figsize=(14, 8))
sns.lineplot(data=mean_stats_by_gen)
plt.title('Tren Statistik Dasar dari Generasi ke Generasi')
plt.xlabel('Generation')
plt.ylabel('Mean Statistik')
plt.legend(stats)
plt.show()
```

###Identifikasi Pokemon Legendary

```python
# Membuat subplots 3x2 untuk perbandingan statistik dasar antara Pokemon Legendary dan non-Legendary
fig, axs = plt.subplots(2, 3, figsize=(18, 12))

for i, stat in enumerate(stats):
    sns.boxplot(x='Legendary', y=stat, data=df, ax=axs[i//3, i%3])
    axs[i//3, i%3].set_title(f'Perbandingan {stat} antara Pokemon Legendary dan non-Legendary')
    axs[i//3, i%3].set_xlabel('Legendary')
    axs[i//3, i%3].set_ylabel(stat)

plt.tight_layout()
plt.show()

# Scatter plot untuk melihat distribusi statistik dasar antara Pokemon Legendary dan non-Legendary
sns.pairplot(df, hue='Legendary', vars=stats)
plt.show()
```
